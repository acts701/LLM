# NLI data set created
- 어찌어찌 chat gpt에 query 보내서 결과를 받긴 했다.
- 1단계 : STT를 기용해서 생성된 text를 google drive에서 읽기
- 2단계 : chat gpt api로 query 보내기
``` python
        prompt = (
            f"다음 텍스트를 분석하여 premise, hypothesis, label(entailment, contradiction, neutral)을 생성해 주세요. "
            f"결과는 {nli_set_count}개의 세트로 작성해 주세요.\n\n텍스트:\n{input_text}"
        )

        # ChatGPT 요청
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo-0125",  # gpt-4 사용 시 모델 변경 가능
            messages=[
                {"role": "system", "content": "You are an expert NLI dataset creator."},
                {"role": "user", "content": prompt}
            ]
        )
```
- 3단계 : premise, hypothesis, label로 나눠서 csv를 google drive에 생성
- 결과
    - 1000000109.jpg
